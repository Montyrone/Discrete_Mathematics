$\textbf{1. {$A \setminus B = A \cap B^C$}}$

$\textbf{Left Side}$
- If $x$ $\in$ $A\setminus B$ then $x$ $\in$ $A$ $\land$ x $\notin$ $B$
  
$\textbf{Right Side}$
- $x$ $\in$ $A$ $\land$ $x$ $\in$ $B^C$ then $x$ $\in$ $A$ $\cap$ $B^C$
- Therefore $A\setminus B$ $\subseteq$ $A$ $\cap$ $B^C$ or $A\setminus B = A$ $\cap$ $B^C$
  
$\textbf{2. $B = (B \cap A) \cup (B \cap A^C)$}$ 

$\textbf{Right Side}$
- If $x$ $\in$ $B$ then $x$ $\in$ ($B$ $\cap$ $A$) $\cup$ ($B$ $\cap$ $A^C$)

- If $x$ $\in$ ($B$ $\cap$ $A$) $\cup$ ($B$ $\cap$ $A^C$) then $x$ $\in$ $B$ $\cap$ ($A$ $\cup$ $A^C$) $\land$ $x$ $\in$ $A$ $\cup A^C$ (Using the Distributive Law)
  
- If $x$ $\in$ $A \cup A^C = \mathcal {U}$ $\land$ $x$ $\in$ $B \cap \mathcal{U}$ = $B$ then $x$ $\in$ $B$ (Using Domination Law)
  
$\textbf{3. $B \setminus A = B \cap A^C$}$ 

$\textbf{Left Side}$
- If $x$ $\in$ $B\setminus A$ then $x$ $\in$ $B$ $\land$ x $\notin$ $A$

$\textbf{Right Side}$
- $x$ $\in$ $B$ $\land$ $x$ $\in$ $A^C$ then $x$ $\in$ $B$ $\cap$ $A^C$
- Therefore $B\setminus A$ $\subseteq$ $B$ $\cap$ $A^C$ or $B\setminus A = B$ $\cap$ $A^C$

$\textbf{4. $A \cup B = A \cup (B\cap A^C$)}$

$\textbf{Left Side}$
- If $x$ $\in$ $A$ $\cup$ $B$ then $A$ $\cup$ $B$ $\subseteq$ $U$

$\textbf{Right Side}$

- If $x$ $\in$ $A$ $\cup$ ($B$ $\cup$ $A^C$) then $x$ $\in$ ($A$ $\cup$ $B$) $\cup$ ($A$ $\cup$ $A^C$) (Using the Distributive Law)
- If $x$ $\in$ ($A$ $\cup$ $B$) $\cup$ $\mathcal{U}$ then $x$ $\in$ $\mathcal{U}$ = $A \cup B$ (Using Domination Law)
- Therefore $A \cup B = A \cup (B\cap A^C$)
  
$\textbf{5.}$ $(\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c}$ = $\cap_{i=1}^{\infty}A_{i}$

$\textbf{Left Side}$
- If $x$ $\in$ $(\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c}$ then $x$ $\notin$ $\cap_{i=1}^{\infty}A_{i}$} $\land$ $x$ $\notin$ $A_i^C$
- If $x$ $\notin$ $A_i^C$ then $x$ $\in$ $A_i$
- If $x$ $\in$ $(\textbf{U}_{i=1}^{\infty} A_{i}^{c})^{c}$ then $x$ $\in$ $A_i$ and for all $\it{i}$ $\land$ $x$ $\in$ $\cap_{i=1}^{\infty}A_{i}$

$\textbf{Right Side}$
- If $x$ $\in$ $\cap_{i=1}^{\infty}A_{i}$ then $x$ $\in$ $A_i$ and for all $\it{i}$

- If $x$ $\notin$ ${U}_{i=1}^{\infty} A_{i}^{c}$ then $x$ $\in$ $({U}_{i=1}^{\infty} A_{i}^{c})^{c}$
- If $x$ $\in$ $\cap_{i=1}^{\infty}A_{i}$ then $x$ $\in$ $({U}_{i=1}^{\infty} A_{i}^{c})^{c}$

Using Demorgan's law which states that the complement of a union is the intersection of the complements, and the complement of an intersection is the union of the complements, we can distribute the complements of both left and right side within the parenthesis and get an x that is present in either sides making both subsets of each other. Meaning that since they are subsets of each other, they also equate to each other.

2. Find the number of people who read at $\textbf{least one}$ of the three magazines.

   $|N \cup T \cup F| = |N| + |T| + |F| − (N\cap T + N\cap F + T\cap F) + |N\cap F\cap T| = 65 + 42 + 20 - (20 + 35 + 15) + 8 = \boxed{100}$
   
  
3. Find the number of people who read exactly one magazine. <br>

-  In order to find the number of people who have read ${ \underline{exactly} }$ one magazine, we have to do the following: 
  
   $|N| = |N| − (N\cap T) - |N\cap F| + |N\cap F\cap T| = 65 - 20 - 25 + 8 = \boxed{28}$

   $|T| = |T| − (N\cap T) - |T\cap F| + |N\cap F\cap T| = 45 - 20 - 15 + 8 = \boxed{18}$

   $|F| = |F| − (N\cap F) - |T\cap F| + |N\cap F\cap T| = 42 - 25 - 15 + 8 = \boxed{10}$
   
   ${ \textbf{Exactly One} }$ = $|N| + |T| + |F| = 28 + 18 + 10$ = $\boxed{56}$
